```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
# Load the CSV file into a DataFrame
data = pd.read_csv('tiempos.csv')



print(data)
```

## Graficos

```python
# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the values
ax.plot(data)
    
# Add labels and title
ax.set_xlabel('Tiempo transcurrido')
ax.set_ylabel('data')
ax.set_title('Mensajes Ping Lineal')

# Show the plot
plt.show()
```
